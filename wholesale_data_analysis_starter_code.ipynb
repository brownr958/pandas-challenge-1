## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["line_subtotal"] = df["unit_price"] * df["qty"]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df["shipping_price"] = df.apply(lambda row: (row["unit_weight"] * row["qty"]) * 7 if (row["unit_weight"] * row["qty"]) > 50 else (row["unit_weight"] * row["qty"]) * 10, axis=1)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["line_price"] = (df["line_subtotal"] + df["shipping_price"]) + ((df["line_subtotal"] + df["shipping_price"]) * 0.0925)
df.head(1)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["line_cost"] = (df["unit_cost"] * df["qty"]) + df["shipping_price"]
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
df["line_profit"] = df["line_price"] - df["line_cost"]
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in top_5_client_ids:
    spend = df["line_price"].loc[df["client_id"] == client].sum()
    print(f'{client}: ${round(spend, 2)}')
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

# Initialize an empty list to store each client's data
client_summary = []

# Loop through each client in the top 5 client IDs
for client in top_5_client_ids:
    # Calculate the required metrics for each client
    qty = df["qty"].loc[df["client_id"] == client].sum()
    shipping_price = df["shipping_price"].loc[df["client_id"] == client].sum()
    line_price = df["line_price"].loc[df["client_id"] == client].sum()
    line_cost = df["line_cost"].loc[df["client_id"] == client].sum()
    line_profit = df["line_profit"].loc[df["client_id"] == client].sum()  # Assuming profit is revenue minus cost
    
    # Append the data for this client as a dictionary
    client_summary.append({
        "client_id": client,
        "total_units_purchased": qty,
        "total_shipping_price": shipping_price,
        "total_revenue": line_price,
        "total_cost": line_cost,
        "total_profit": line_profit
    })

# Create a DataFrame from the list of dictionaries
topclient_df = pd.DataFrame(client_summary)

# Display the summary DataFrame
topclient_df

```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_columns = topclient_df[["total_shipping_price","total_revenue","total_cost","total_profit"]]

# Define a function that converts a dollar amount to millions.
def millions(column):
    return column / 1000000 

# Apply the currency_format_millions function to only the money columns. 
for column in money_columns:
    topclient_df[column] = topclient_df[column].apply(millions)

# Rename the columns to reflect the change in the money format. 
topclient_df = topclient_df.rename(columns={
    "client_id": "Client ID",
    "total_units_purchased": "Units",
    "total_shipping_price": "Shipping (millions)",
    "total_revenue": "Total Revenue (millions)",
    "total_cost": "Total Cost (millions)",
    "total_profit": "Total Profit (millions)"
})

topclient_df.head(5)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
topclient_df.sort_values(by="Total Profit (millions)", ascending=False).head(5)
```

The top 5 clients contributed significantly to the company's revenue, with total revenue ranging in the millions. Client 24741 was the most profitable, indicating a high engagement with premium products. These insights can guide targeted marketing and retention efforts for top clients.
